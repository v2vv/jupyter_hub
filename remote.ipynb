# docker run

-i -i: 以交互模式运行容器，通常与 -t 同时使用；  

-t: 为容器重新分配一个伪输入终端，通常与 -i 同时使用；  

--name "nginx-lb": 为容器指定一个名称；  

-d: 后台运行容器，并返回容器ID；  

-p: 指定端口映射，格式为：主机(宿主)端口:容器端口  

--volume , -v: 绑定一个卷  

--rm 在容器退出后，删除容器，并且会删除挂载的数据卷，主机映射目录的 目录不会被删除，而没有指定的映射目录则会被删除
